```python
# importing Libraries
import seaborn as sns
# import pandas as pd
import pandas as pd

# importing matplotlib module
import matplotlib.pyplot as plt
plt.style.use('default')

# %matplotlib inline: only draw static
# images in the notebook
%matplotlib inline
```

```python
# importing Data
banjir = pd.read_csv('banjir.csv')
banjir.head(78)
```

# **Aggrigate Data**

1.   Total Floods in  each Year

```python
Bil_banjir = banjir.groupby(["Tahun"])["Tahun"].count().reset_index(name="Bil Banjir")
Bil_banjir.head(78)
```

```python
# Save as newsdatastudent1.csv
Bil_banjir.to_csv("Bil_banjir.csv")
```

2.   Total Floods that happen in each State

```python
total_by_negeri = banjir.groupby("negeri")["tempat"].count().reset_index(name="Bil Kes")
total_by_negeri
```

3.   Total Floods that happen in each State and year

```python
Bil_Banjir = banjir.groupby(["tempat","Tahun"])["tempat"].count().reset_index(name="Bil Banjir")
Bil_Banjir.head(78)
```

# **Visualize Data**


1. Bar Graph to show the Total Floods that happen in each State

```python
sns.set(style="whitegrid", color_codes=True)
sns.barplot(x = 'negeri', y = 'Bil Kes', data=total_by_negeri)
plt.xticks(rotation = 90)
plt.show()
```

2. Bar Graph to show the Total Floods in each Year

```python
sns.set(style="whitegrid", color_codes=True)
sns.barplot(x = 'Tahun', y = 'Bil Banjir', data=Bil_banjir)
plt.xticks(rotation = 90)
plt.show()
```

3. Total Floods that happen in Malaysia using Line graph

```python
# code
# importing Data
Bil_banjir = pd.read_csv('Bil_banjir.csv',
					index_col='Tahun',
					parse_dates=True)
Bil_banjir.head(78)
```

```python
# code
# Visualizing Floods that happen in Malaysia using Line graph

# to set the plot size
plt.figure(figsize=(16, 8), dpi=150)

# using plot method to plot data.
# in plot method we set the label and color of the curve.
Bil_banjir['Bil Banjir'].plot(label='banjir', color='orange')


# adding title to the plot
plt.title('Floods that happen in Malaysia using Line graph')

# adding Label to the x-axis
plt.xlabel('Tahun')

# adding legend to the curve
plt.legend()
```
